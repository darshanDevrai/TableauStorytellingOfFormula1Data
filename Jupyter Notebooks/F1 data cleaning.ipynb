```python
import pandas as pd
```

```python
df = pd.read_csv('results.csv')
```

```python
df.info()
```

```python
new_points = []
def calculateThePoints(row):
    if row['positionText'] == "1":
        new_points.append(25)
        return
    elif row['positionText'] == "2":
        new_points.append(18)
        return
    elif row['positionText'] == "3":
        new_points.append(15)
        return
    elif row['positionText'] == "4":
        new_points.append(12)
        return
    elif row['positionText'] == "5":
        new_points.append(10)
        return
    elif row['positionText'] == "6":
        new_points.append(8)
        return
    elif row['positionText'] == "7":
        new_points.append(6)
        return
    elif row['positionText'] == "8":
        new_points.append(4)
        return
    elif row['positionText'] == "9":
        new_points.append(2)
        return
    elif row['positionText'] == "10":
        new_points.append(1)
        return
    else:
        new_points.append(0)
        return

for index, row in df.iterrows():
    calculateThePoints(row)
    
    
df['new_points'] = new_points  
    
```
